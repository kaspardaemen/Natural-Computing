```python
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
```

```python
def plotfunc(seq, locs):
    sortlocs = [locs[_ind] for _ind in seq]
    xs = [loc[0] for loc in sortlocs]
    ys = [loc[1] for loc in sortlocs]
    
    plt.figure(0)
    plt.scatter(xs, ys)
    plt.plot(xs, ys)
    plt.show()
    
```

```python
locs = np.loadtxt('file-tsp.txt')
```

```python
np.random.seed(2)
seq = random.permutation([i for i in range(0,50)])
# seq = [i for i in range(0,50)]
```

```python
plotfunc(seq, locs)
```
