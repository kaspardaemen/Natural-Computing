## c. Explain what is the effect of the parameter ω.

The $\omega$ parameter determines the influence of the velocity of the previous
iteration on the calculation of the velocity of the current iteration. Basically,
if $\omega$ is large, the influence of the previous velocity is large as well.
One could think of it as a measure of how heavy the particle is if it would be
a physical particle. The larger the $\omega$, the heavier the particle,
and therefore the more energy is needed to change its course when choosing to go another direction.
If $\omega$ is smaller it becomes easier to change course, since e.g. braking on a bicycle
also is easier than braking when you are driving a pick-up truck with a loaded trailer.

## d. Give an advantage and a disadvantage of a high value of ω.

- The advantage of a large $\omega$ lies in the ability to make larger steps per iteration.
This allows for a more diverse search of the solution-space as larger leaps can be made
and therefore solutions that are more radically different can be evaluated.
- A disadvantage is the larger probability of overshooting the target. A situation
in which the current iteration has determined that the target lies in the other direction
than the velocity of the previous iteration, the large $\omega$ will prevent a
large enough change in velocity to actually steer in the right direction. Instead it will
continue in the wrong direction for maybe even more than one iteration, which is
not optimal.
