## c. Explain what is the effect of the parameter œâ.

The $\omega$ parameter determines the influence of the velocity of the previous
iteration on the calculation of the velocity of the current iteration. Basically,
if $\omega$ is large, the influence of the previous velocity is large as well.
One could think of it as a measure of how heavy the particle is if it would be
a physical particle. The larger the $\omega$, the heavier the particle,
and therefore the more energy is needed to change its course when choosing to go another direction.
If $\omega$ is smaller it becomes easier to change course, since e.g. braking on a bicycle
also is easier than braking when you are driving a pick-up truck with a loaded trailer.

## d. Give an advantage and a disadvantage of a high value of œâ.

- The advantage of a large $\omega$ lies in the ability to make larger steps per iteration.
This allows for a more diverse search of the solution-space as larger leaps can be made
and therefore solutions that are more radically different can be evaluated.
- A disadvantage is the larger probability of overshooting the target. A situation
in which the current iteration has determined that the target lies in the other direction
than the velocity of the previous iteration, the large $\omega$ will prevent a
large enough change in velocity to actually steer in the right direction. Instead it will
continue in the wrong direction for maybe even more than one iteration, which is
not optimal.

## Consider a particle ‚Äúswarm‚Äù consisting of a single member. ow would it perform in a trivial task such as the minimization of f(x) = $x^2$ when œâ < 1, ssuming the particle starts with the velocity pointing away from the optimum (e.g., in a state with velocity v = 10; position x = 20)?

The algorithm would need some time to reverse the direction of the velocity so that it is headed towards the optimal value (a negative velocity that is headed towards 0, in the state example of the question).

But this will happen very fast within a few epochs since the difference between the resulting value (resulting value of x^2) of the best particle at the beginning and the resulting value of the actual position of the particle will increase very rapidly since the resulting value of x^2 grows exponentially.

When the function would approach zero, the velocity would decrease very fast, due to the fact that the difference between the best solution of the particle (and the global best solution) and the current solution would be very small near to the optimal solution. Therefore, it would converge very fast around the optimal value of 0. 

So the algorithm would perform very good for a trivial task such as the minimization of f(x) =  ùë•^2.

We have implemented the state example of the question in python code below, where the algorithm runs for 20 epochs. A plot of the resulting value of the particle for each epoch is shown below the code, visualizing the performance of the algorithm. 

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

class PSO ():
    
    def __init__ (self, x,  v, w, a, b):
        self.x = x
        self.v = v
        self.w = w
        self.a = a
        self.b = b
        self.best_x = x.copy()
        self.g = x[0]
        self.hist = []

    def f (self, x):
        return x**2
    
    def get_fitness (self):
        for i in range(len(self.x)):
            fitness = self.f(self.x[i])
            x = self.x[i]
            v = self.v[i]
    
    def plot_fitness (self):
        x = [x for x in range(len(self.hist))]
        y = [x for x in self.hist]
        plt.xlabel('Epoch')
        plt.ylabel('Resulting value (lower is better)')
        plt.title('Plot of the fitness of a swarm with only one particle, initialised with the velocity pointing in the wrong direction')
        plt.plot(x,y)
        plt.plot()
        
    
    def update(self):
        r1 = 0.5
        r2 = 0.5
        
        #update velocities
        for i in range(len(self.v)):
            self.hist.append(self.f(self.x[i]))
            term1 = self.w * self.v[i] 
            term2 =  self.a * r1  * (self.best_x[i] - self.x[i])
            term3 = self.a * r2  * (self.best_x[i] - self.x[i])
            self.v[i] = term1 + term2 + term3
          
         #update positions and global best
        for i in range(len(self.v)):
            self.x[i] = self.x[i] + self.v[i]
            #print(self.x[i], self.best_x[i])
            self.best_x[i] = self.x[i] if (self.f(self.x[i]) < self.f(self.best_x[i])) else self.best_x[i]
            self.g = self.x[i] if (self.f(self.x[i]) < self.f(self.g)) else self.g
    def run(self):
        for i in range(20):
            self.get_fitness()
            self.update()
```

```python
#particle:
x = [20]
#weight:
w = 0.8
#alpha_1 and alpha_2:
a = 1
b = 1

#vecolicity
v=[10]

#Run the PSO algorithm:
swarm = PSO(x=x, w=w, a=a, b=b, v=v)
swarm.run()
swarm.plot_fitness()
```

## a. Is ACO for this problem a Competition-Balanced System (CBS)? Justify your answer. (A definition of CBS also given in the slides.)


ACO for this problem is not a competition balanced system. This is because not all c in C where C is the set of solution components occur in the same number of feasible solutions.

A feasible solution is described by Blum, Christian, and Marco Dorigo as "A feasible solution is a complete assignment (i.e., an assignment in which each decision variable has a domain value assigned) that satisfies the constraints". 

A solution for this problem consists of 2 variables: edge A and edge B which both have a domain value assigned to them (the edge weight. 

In figure 1, we have three solutions: 1-2-3 with a weight of 3, 2-3-4 with a weight of 4 and 3-4-5 with a weight of 3. All of these solutions are feasible solutions, because they all have a value that falls within the domain which is the domain fo the natural numbers for this problem. 

For any feasible partial solution we take, we can form one of these three solutions, since a partial solution consists of one edge, and we are allowed to add one more edge to complete the solution. 

In the slides, a Competition-balanced system is described as: "An ACO and a problem instance are Competition Balanced System (CBS) if given a feasible partial solution Sp and the set C of solution components that can be added to extend Sp, all c in C occur in the same number of feasible solutions"

As we can see, 3 occurs in 3 solutions, 2 and 4 occur in 2 solutions and 1 and 5 occur in one solution. Thus not all c occur in the same number of feasible solutions, and thus the problem is not competition balanced. 

## b. If a combination of an ACO algorithm and a problem instance is not a CBS, is the induced bias always harmful? Justify your answer.

Not always, I would argue that the probabilistic nature of ACO would prevent bias to always be harmful. If the problem instance has a small enough solution space such that there is a difference but not a large one for some solution components to occur in feasible solutions compared to other solution components, there is bias but there is still a large enough probability for all solution components to be attempted. Because if those solutions are attempted and they turn out to be adequate solutions, the appropriate pheromones will be released and result in more extensive investigation of that solution. And because this will signal more ants towards this solution, the bias will be minimal, meaning that the bias is not always a harmful aspect.
